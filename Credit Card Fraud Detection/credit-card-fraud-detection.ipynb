Descriptive Statistics for Fraudulent Transactions Amount

```python
fraudulent_amount_stats = data[data['Class'] == 1]['Amount'].describe().round(2)
print("Descriptive Statistics for Fraudulent Transactions Amount:")
print(fraudulent_amount_stats)
```

```python
X = data.drop(columns=['Class'], axis=1)  
y = data['Class']
```

**Splitting the Data into Training and Test Sets**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

**Standardizing the Explanatory Variables**

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**SVM Model **

```python
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

**Evaluation**

```python
train_accuracy = model_svc.score(X_train, y_train)
test_accuracy = model_svc.score(X_test, y_test)
print("Accuracy on training set:", train_accuracy)
print("Accuracy on test set:", test_accuracy)
```

**Predictions**

```python
y_predict = model_svc.predict(X_test)
```

**Classification Report and Confusion Matrix**

```python
print(classification_report(y_test, y_predict))
confusion = confusion_matrix(y_test, y_predict)
sns.heatmap(confusion, annot=True, fmt='g', cmap='Blues', 
            xticklabels=['Genuine', 'Fraud'], yticklabels=['Genuine', 'Fraud'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```
